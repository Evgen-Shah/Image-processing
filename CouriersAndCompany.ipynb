```python
!python /content/drive/MyDrive/contentCourier/yolov5/detect.py --weights /content/drive/MyDrive/contentCourier/yolov5/runs/train/exp2/weights/last.pt --img 640 --conf 0.4 --source /content/drive/MyDrive/contentCourier/yolov5/data/images --save-txt
```

```python
path="drive/MyDrive/contentCourier/yolov5/runs/detect/exp8"
pathTxt="drive/MyDrive/contentCourier/yolov5/runs/detect/exp8/labels"
for i in range(1,9):
  im = Image.open(path+'/'+str(i)+'.jpg')
  im = np.array(im)
  f = open(pathTxt + '/'+ str(i) +'.txt',"r")
  dh,dw, _ = im.shape
  for string in f:
         l, x, y, w, h = map(float, string.split(' '))
         x = int((x - w / 2) * dw)
         w = int(w * dw)
         y = int((y - h / 2) * dh)
         h = int(h * dh)
         if(l == 3):
            cv2.rectangle(im, (x,y), (x + 220,y - 40), (0,0,0), -1)
            cv2.putText(im, str('8(800) 505-00-15'), (x,y - 15),cv2.FONT_HERSHEY_SIMPLEX,0.7, (255, 255, 255), 1,cv2.LINE_AA  )
         elif (l == 2):
            cv2.rectangle(im, (x,y), (x + 220,y - 40), (0,0,0), -1)
            cv2.putText(im, str('8(800) 770-75-09'), (x,y - 15),cv2.FONT_HERSHEY_SIMPLEX,0.7, (255, 255, 255), 1,cv2.LINE_AA  )
         elif (l == 1):
            cv2.rectangle(im, (x,y), (x + 220,y - 40), (0,0,0), -1)
            cv2.putText(im, str('8(812) 209-00-25'), (x,y - 15),cv2.FONT_HERSHEY_SIMPLEX,0.7, (255, 255, 255), 1,cv2.LINE_AA  )
         elif (l == 0):
            cv2.rectangle(im, (x - 10,y), (x + 260,y - 40), (0,0,0), -1)
            cv2.putText(im, str('8(800) 600-12-10'), (x-12,y - 15),cv2.FONT_HERSHEY_SIMPLEX,0.7, (255, 255, 255), 1,cv2.LINE_AA  )
  im = cv2.cvtColor(im,cv2.COLOR_BGR2RGB)
  cv2_imshow(im)

```
