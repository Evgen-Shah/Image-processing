```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

path="/content/drive/MyDrive/Money"
ref = Image.open(path+'/mn2.jpg')
ref = np.array(ref)
ref = cv2.cvtColor(ref,cv2.COLOR_BGR2RGB)
hR = ref.shape[0]
wR = ref.shape[1]
copyimg = ref.copy()
total = 0
ref_grey = cv2.cvtColor(ref,cv2.COLOR_BGR2GRAY)

path="/content/drive/MyDrive/Money"
for image in range(1,5):
  template = Image.open(path+'/' + str(image) +'.jpg')
  template = np.array(template)
  template = cv2.cvtColor(template,cv2.COLOR_BGR2RGB)

  cv2_imshow(template)

  h = template.shape[0]
  w = template.shape[1]
  mask = np.zeros((hR+2, wR+2), np.uint8) 
  res = cv.matchTemplate(ref,template,cv.TM_CCOEFF_NORMED)


  threshold = 0.5
  loc = np.where( res >= threshold)
  for pt in zip(*loc[::-1]): #–≤—Å–µ –Ω–∞–π–¥–µ–Ω–Ω—ãe —Ç–æ—á–∫–∏ –ø–æ —Ç–µ–º–ø–ª–µ–π—Ç—É, –ø–µ—Ä–µ–±–æ—Ä
    if mask[pt[1] + int(round(h/2)), pt[0] + int(round(w/2))] != 255: #—á—Ç–æ–±—ã –ø–æ–≤—Ç–æ—Ä–Ω—ã–µ –Ω–µ –≤—ã–¥–µ–ª—è–ª–∏–ª—Å—å
          mask[pt[1]:pt[1]+h, pt[0]:pt[0]+w] = 255
          if(image == 1):
            total += 1
          elif (image == 2):
            total +=2
          elif (image == 3):
            total+= 5
          elif (image == 4):
            total+= 10
          cv2.rectangle(ref, pt, (pt[0] + w, pt[1] + h), (0,255,255), 3)

cv2_imshow(ref)
print('–†—É–±–ª–µ–π –≤—Å–µ–≥–æ: ' + str(total))
```

```python
def deleteDuplicate(mn,x,y,arrayX,arrayY,mnS):
 # print(len(arrayX))
  if(len(arrayX) == 0):
    arrayX.append(x)
    arrayY.append(y)
    mnS.append(mn)
    return True

  else:
    arrayX.append(x)
    arrayY.append(y)
    mnS.append(mn)
    for i in range(0,len(arrayX)):
      if(i == (len(arrayX) - 1)):
        return True
      if(abs(x - arrayX[i]) < 60 and abs(y - arrayY[i]) < 60 and mn == mnS[i]) :
        return False
```
